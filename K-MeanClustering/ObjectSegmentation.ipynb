```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg # thư viện dùng để đọc ảnh
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Đọc file ảnh
img=mpimg.imread('girl.jpg')
# hiển thị ảnh
#plt.imshow(img)
# tạo đổi tượng AxesImage
imgplot=plt.imshow(img)

# Mỗi điểm ảnh gồm 3 số
# tạo X là tập hợp các điểm ảnh
X=img.reshape(img.shape[0]*img.shape[1], img.shape[2])

```

```python
for K in [5, 10]:
    kmeans=KMeans(n_clusters=K).fit(X)
    # mảng chỉ số của cluster tương ứng với mỗi điểm ảnh
    label=kmeans.predict(X)
    # tạo ma trận không có chiều giống vs X
    img4=np.zeros_like(X)
    # thay thế các điểm ảnh bằng center
    for k in range(K):
        img4[label==k]=kmeans.cluster_centers_[k]
        #reshape and display
    img5=img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
    
```
